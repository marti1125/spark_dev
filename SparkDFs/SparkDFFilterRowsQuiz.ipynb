```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
spark = SparkSession.builder.appName("SparkDFFilterRowsQuiz").getOrCreate()
```

```python
df = spark.read.options(header="True", inferSchema="True").csv("StudentData.csv")
df.show()
```

```python
df = df.withColumn("total_marks", lit(120))
df.show()
```

```python
df = df.withColumn("average", (col("marks") / col("total_marks") * 100))
df.show()
```

```python
df_OOP = df.filter((col("course") == "OOP") & (col("average") > 80))
df_OOP.show()
```

```python
df_Cloud = df.filter((col("course") == "Cloud") & (col("average") > 60))
df_Cloud.show()
```

```python
df_OOP.select("name", "marks").show()
```

```python
df_Cloud.select("name", "marks").show()
```
