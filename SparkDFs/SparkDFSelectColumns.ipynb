```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('SparkDFSelectColumns').getOrCreate()
```

```python
df = spark.read.options(header="True", inferSchema="True").csv("StudentData.csv")
```

```python
df.printSchema()
df.show()
```

```python
df.select("name", "gender").show()
```

```python
df.select(df.name, df.gender).show()
```

```python
from pyspark.sql.functions import col
```

```python
df.select(col("roll"), col("name")).show()
```

```python
df.select("*").show()
```
