```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setAppName("Read File")
```

```python
sc = SparkContext.getOrCreate(conf)
```

```python
text = sc.textFile("sample.txt")
```

```python
text.collect()
```

```python
l2 = text.map(lambda x: [int(a) for a in x.split(" ")])
l2.collect()
```

```python
def arr(x):
    return [int(a) ** 2 for a in x.split(" ")]
```

```python
l3 = text.map(arr)
l3.collect()
```

```python
text2 = sc.textFile("sample_text.txt")
```

```python
l3 = text2.flatMap(lambda x: [len(a) for a in x.split(" ")])
l3.collect()
```

```python
l4 = text2.filter(lambda x: x != "great")
l4.collect()
```

```python
text3 = sc.textFile("phrases.txt")
```

```python
l5 = text3.filter(lambda x: not (x.startswith("a") or x.startswith("c")))
l5.collect()
```
